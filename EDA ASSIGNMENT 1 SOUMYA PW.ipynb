## Q1. What is the average, median, and standard deviation of house prices?

```python
average_price = df['Price'].mean()
median_price = df['Price'].median()
std_price = df['Price'].std()

print("Average Price:", average_price)
print("Median Price:", median_price)
print("Standard Deviation:", std_price)
```

Interpretation-:
-If standard deviation is high, prices vary greatly.
-If mean > median, dataset may contain expensive outliers.

## Q2. Which number of bedrooms is most common?

```python
most_common_bedrooms = df['number of bedrooms'].mode()[0]

print("Most Common Bedrooms:", most_common_bedrooms)
```

Interpretation-: Shows the most demanded house configuration.

## Q3. Check if the price data is skewed.

```python
price_skewness = df['Price'].skew()

print("Price Skewness:", price_skewness)
```

```python
sns.histplot(df['Price'], kde=True)
plt.title("Price Distribution")
plt.show()
```

Interpretation:-
--Skewness > 0 → Right-skewed
--Skewness < 0 → Left-skewed
--Skewness ≈ 0 → Normal distribution

Usually house prices are positively skewed because few luxury houses have very high prices.

## Q4. What is the average price for each number of bedrooms?

```python
avg_price_bedrooms = df.groupby('number of bedrooms')['Price'].mean()

print(avg_price_bedrooms)
```

```python
avg_price_bedrooms.plot(kind='bar')
plt.title("Average Price by Bedrooms")
plt.xlabel("Bedrooms")
plt.ylabel("Average Price")
plt.show()
```

Interpretation-: More bedrooms generally increase price.

## Q5. What is the relationship between living area and price?

```python
sns.scatterplot(x=df['living area'], y=df['Price'])

plt.title("Living Area vs Price")
plt.xlabel("Living Area")
plt.ylabel("Price")
plt.show()
```

```python
correlation = df['living area'].corr(df['Price'])

print("Correlation:", correlation)
```

Interpretation -: Positive correlation means:
                 Larger houses usually cost more.

If correlation is near 1: Strong relationship exists.

## Q6. Identify anomalies where houses have high prices but low areas.

```python
anomalies = df[
    (df['Price'] > df['Price'].quantile(0.90)) &
    (df['living area'] < df['living area'].quantile(0.25))
]

print(anomalies[['Price', 'living area']])
```

Interpretation

Possible reasons:
-Prime location
-Luxury interiors
-Waterfront view

## Q7. Compare average price based on number of floors and houses with or without waterfront.

```python
comparison = df.groupby(['number of floors', 'waterfront present'])['Price'].mean()

print(comparison)
```

```python
comparison.unstack().plot(kind='bar')

plt.title("Average Price by Floors and Waterfront")
plt.ylabel("Average Price")
plt.show()
```

Interpretation-:
Waterfront properties usually cost more.
Multi-floor houses may have higher value.

## Q8. Identify the minimum and maximum house price. What does this indicate?

```python
min_price = df['Price'].min()
max_price = df['Price'].max()

print("Minimum Price:", min_price)
print("Maximum Price:", max_price)
```

Interpretation-:
-Large difference indicates:
    --Wide market variation
    --Presence of luxury and budget houses

## Q9. Which location (zipcode/area) has the highest average price?

```python
highest_location = df.groupby('Postal Code')['Price'].mean().sort_values(ascending=False)

print(highest_location.head(1))
```

Interpretation-:
Indicates premium residential areas.
