```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\soumy\Downloads\Titanic.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

* Q1. Create a countplot to show the number of passengers by gender

```python
plt.figure(figsize=(6,4))

sns.countplot(x='Sex', data=df, palette='Set2')

plt.title("Passengers by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.show()
```

Male passengers are more than female passengers.

* Q2. Create a bar chart to show passengers in each class

```python
plt.figure(figsize=(6,4))

sns.countplot(x='Pclass', data=df, palette='viridis')

plt.title("Passengers in Each Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")

plt.show()
```

Most passengers traveled in 3rd class.

* Q3. Create a histogram for passenger age distribution

```python
plt.figure(figsize=(8,5))

sns.histplot(df['Age'], bins=30, kde=True, color='skyblue')

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

Most passengers were between 20–40 years old.

* Q4. Create a boxplot to detect outliers in Fare column

```python
plt.figure(figsize=(8,4))

sns.boxplot(x=df['Fare'], color='orange')

plt.title("Fare Boxplot")

plt.show()
```

Many outliers are present in the Fare column.

* Q5(a). Scatterplot between Age and Fare

```python
plt.figure(figsize=(8,5))

sns.scatterplot(x='Age', y='Fare', data=df)

plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")

plt.show()
```

Passengers paying higher fares belong mostly to higher classes.

* Q5(b). Relationship between Family Size and Fare

```python
df['FamilySize'] = df['SibSp'] + df['Parch']

plt.figure(figsize=(8,5))

sns.scatterplot(x='FamilySize', y='Fare', data=df, color='red')

plt.title("Family Size vs Fare")
plt.xlabel("Family Size")
plt.ylabel("Fare")

plt.show()
```

Some large families paid higher fares.

* Q6. Create a countplot for survival by gender

```python
plt.figure(figsize=(6,4))

sns.countplot(x='Sex', hue='Survived', data=df)

plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.show()
```

Female passengers survived more than male passengers.

* Q7. Create a bar chart for survival by passenger class

```python
plt.figure(figsize=(6,4))

sns.barplot(x='Pclass', y='Survived', data=df)

plt.title("Survival by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")

plt.show()
```

1st class passengers had the highest survival rate.

* Q8. Create a heatmap for correlation

```python
plt.figure(figsize=(10,6))

sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap='coolwarm')

plt.title("Correlation Heatmap")

plt.show()
```

Interpretation
--Fare and Survival are positively related.
--Pclass and Fare are negatively related.

* Q9. Identify which passenger class paid the highest average fare

```python
avg_fare = df.groupby('Pclass')['Fare'].mean()

print(avg_fare)
```

1st class passengers paid the highest average fare.
